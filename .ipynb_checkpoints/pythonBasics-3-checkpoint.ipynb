```python
#71 iterables
#In Python  list,sictionary,tuple,set and string sre iterable
#iterable means simply an object or caollection the can be iterated over

#iterating dictionary
users={
    'name':'Ajay',
    'age':45,
    'can_swim': True
}


for item in users:
    print(item)

#dictionary has three main methods when we wants to loop over their key value pair
#1- .items() returns key, value pairs in tuple
for item in users.items():
    print(item)
#2- .values()  returns the values of dictionary
for item in users.values():
    print(item)
#3- keys
for item in users.keys():
    print(item)

#Tuple unpacking
for item in users.items():
    key,value=item
    print(key,value)

#we can write the above code as 

for key,values in users.items():
    #key,value=item
    print(key,value)
```

```python
#73 Range
#range is a special type of object , return an object that produce a sequence of object from start to stop : range(start,stop,stepover)
print(range(100))  
print(range(0,100))

for num in range(0,5):
    print(num)

for _ in range(0,5):
    print(_)

for num in range(0,10,2):
    print(num)

for num in range(10,0,-2):
    print(num)
    
for _ in range(0,10,2):
    print(list(range(10)))

    
```

```python
#74 Enumarator
# enumerate : return an enumerate object and take iterable as an argument
for i,char in enumerate("Hellloooo"):
    print(i,char)

for i,char in enumerate(range(100)):
    if(char==50):
        print(f'index of 50 is -{i}')
```
