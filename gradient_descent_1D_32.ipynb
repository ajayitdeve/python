```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
#function (as a function)
def fx(x):
  return 3*x**2-3*x+4

#derivative function
def deriv(x):
  return 6*x-3
#plot the function and its derivative
x=np.linspace(-2,2,2001)
#plotting
plt.plot(x,fx(x),x,deriv(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```
