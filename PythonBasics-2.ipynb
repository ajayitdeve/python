```python
#57 tuple
# tuple is just like lists but immutable, We can think tuple is an immutable list
my_tuple=(5,6,7)
print(my_tuple)
print(my_tuple[1])
print(5 in my_tuple)

# a tuple can be key of a dictionary

my_dict={
    my_tuple:['a','b','c'],
}
print(my_dict.get(my_tuple))

x=my_tuple[0]
y=my_tuple[1]

print(x )
print(y)
```
